```python
edit_cols = ['Native_country','Occupation','Workclass']
# Replace ? with Unknown
for col in edit_cols:
    data.loc[data[col] == '?', col] = 'unknown'
```

```python
# Check if ? is present
for col in edit_cols:
    print(f"? in {col}: {data[(data[col] == '?')].any().sum()}")
```

```python
hs_grad = ['HS-grad','11th','10th','9th','12th']
elementary = ['1st-4th','5th-6th','7th-8th']

# replace elements in list.
data['Education'].replace(to_replace = hs_grad,value = 'HS-grad',inplace = True)
data['Education'].replace(to_replace = elementary,value = 'elementary_school',inplace = True)

data['Education'].value_counts()
```

```python
married= ['Married-spouse-absent','Married-civ-spouse','Married-AF-spouse']
separated = ['Separated','Divorced']

#replace elements in list.
data['Marital_status'].replace(to_replace = married ,value = 'Married',inplace = True)
data['Marital_status'].replace(to_replace = separated,value = 'Separated',inplace = True)

data['Marital_status'].value_counts()
```

```python
self_employed = ['Self-emp-not-inc','Self-emp-inc']
govt_employees = ['Local-gov','State-gov','Federal-gov']

#replace elements in list.
data['Workclass'].replace(to_replace = self_employed ,value = 'Self_employed',inplace = True)
data['Workclass'].replace(to_replace = govt_employees,value = 'Govt_employees',inplace = True)
```

```python
X = X.drop(['Workclass', 'Education', 'Race', 'Sex',
            'Capital_loss', 'Native_country'], axis=1)
```

```python
#Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
for col in X.columns:
    scaler = StandardScaler()
    X[col] = scaler.fit_transform(X[col].values.reshape(-1, 1))
```

```python
# Fixing imbalanced dataset using Oversampling
```

```python
round(Y.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
```

```python
ros.fit(X, Y)
```

```python
X_resampled, Y_resampled = ros.fit_resample(X, Y)
```

```python
round(Y_resampled.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
# Creating a train test split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_resampled, Y_resampled, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", Y_train.shape)
print("Y_test shape:", Y_test.shape)
```
